```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('iris.csv')
```

```python
df.sample(5)
```

```python
df['Species']=df['Species'].replace({'Iris-versicolor':1,'Iris-setosa':2,'Iris-virginica':3})
df.sample(10)
```

```python

plt.figure(figsize=(8,7))
plt.scatter(df['PetalLengthCm'],df['SepalLengthCm'],c=df['Species'],cmap='jet') #cmap different colors
plt.colorbar() #colorbar at the side
```

```python
batters = pd.read_csv('batter.csv')
```

```python
batters.shape
```

```python
sample_df = batters.head(100).sample(25,random_state=5)
```

```python
sample_df.shape
```

```python
sample_df.head()
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
#Horizontal and vertical lines

plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

plt.axhline(130,color='red')
plt.axhline(140,color='green')
plt.axvline(30,color='red')

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
#Subplots

fig,ax=plt.subplots(nrows=2,ncols=1)
ax[0].scatter(df['SepalLengthCm'],df['SepalWidthCm'],c=df['Species'], cmap='jet')
ax[1].scatter(df['PetalLengthCm'],df['PetalWidthCm'],c=df['Species'], cmap='jet')
ax[0].set_title('Iris graph')
ax[0].set_xlabel('SepalLengthCm')
ax[0].set_ylabel('SepalWidthCm')
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(batters['avg'],batters['strike_rate'])
ax[0,1].scatter(batters['avg'],batters['runs'],color='green')
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'],color='green')
```

```python
#3d plots

fig=plt.figure()
ax=plt.subplot(projection='3d')
ax.scatter3D(df['SepalLengthCm'],df['SepalWidthCm'],df['PetalWidthCm'], color='red', marker='*')
ax.set_title('Iris flowers')

ax.set_xlabel('SepalLengthCm')
ax.set_ylabel('SepalWidthCm')
ax.set_zlabel('PetalWidthCm')
```

```python
#surfaceplot

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(10,8))

ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
