```python
def find_distinct_values(dataset, columns):
    distinct_value_counts = {}

    for column in columns:
        distinct_values = dataset[column].value_counts()
        distinct_value_counts[column] = distinct_values

    return distinct_value_counts
```

```python
# Specify the columns for which you want to find distinct values
columns_to_check = ['gender', 'Topic','Relation']

# Call the function
distinct_values = find_distinct_values(df, columns_to_check)

# Print the distinct values for each column
for column, values in distinct_values.items():
    print(f"Distinct values in {column}:")
    print(values)
    
    print()
```

```python
def count_distinct_values(dataset):
    distinct_value_counts = {}

    for column in dataset.columns:
        distinct_values = dataset[column].nunique()
        distinct_value_counts[column] = distinct_values

    return distinct_value_counts
```

```python
count_distinct_values(df)
```

# OneHotEncoding using Sklearn

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:9],df.iloc[:,-1],test_size=0.2,random_state=91)
```

```python
from sklearn.preprocessing import OneHotEncoder

```

```python
oe=OneHotEncoder(drop='first',sparse=False)
```

```python
X_train_new =oe.fit_transform(X_train[['gender','Topic','Relation']])
```

```python
X_train_new
```

```python
X_test_new =oe.transform(X_test[['gender','Topic','Relation']])
```

```python
X_test_new
```

```python
#convert back to dataframe
X_train_new = pd.DataFrame(X_train_new, columns=oe.get_feature_names(['gender','Topic','Relation']))
```

```python
X_train_new
```

```python
X_test_new = pd.DataFrame(X_test_new, columns=oe.get_feature_names(['gender','Topic','Relation']))
```

```python
X_test_new
```

```python
columns_to_drop = ['gender','Topic','Relation']
```

```python
X_train_dropped = X_train.drop(columns_to_drop, axis=1)
X_test_dropped = X_test.drop(columns_to_drop, axis=1)
```

```python
# Reset the indexes of both dataframes
X_train_dropped.reset_index(drop=True, inplace=True)
X_train_new.reset_index(drop=True, inplace=True)
X_test_dropped.reset_index(drop=True, inplace=True)
X_test_new.reset_index(drop=True, inplace=True)
```

```python
X_train_combined = pd.concat([X_train_dropped, X_train_new], axis=1)
X_test_combined = pd.concat([X_test_dropped, X_test_new], axis=1)
```
