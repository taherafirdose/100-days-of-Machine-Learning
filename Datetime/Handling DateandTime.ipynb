```python
df['today'] = pd.to_datetime("today")
df['td']=df['today'] - df['time']
```

```python
df.sample(5)
```

```python
df['td_in_sec']=(df['today'] - df['time'])/np.timedelta64(1,'s')
df.sample(5)
```

```python
# Convert time zone
df['newtime'] = df['time'].dt.tz_localize('UTC').dt.tz_convert('US/Eastern')

df[['time','newtime']]
```

```python
#Adding a time interval to a datetime column:
df['new_date'] = df['time'] + pd.Timedelta(days=7)
df[['time','new_date']].sample(5)
```

```python
#Adding a specific number of hours to a datetime column

df['hoursadded'] = df['time'] + pd.DateOffset(hours=3)
df[['time','hoursadded']].sample(5)
```

```python
df['yearselapsed'] = (df['today'] - df['time']).dt.days // 365

df[['time','today','yearselapsed']].sample(5)
```

```python
#Adding a specific number of business days to a datetime column:
df['daysadded'] = df['time'] + pd.offsets.BDay(2)
df[['time','daysadded']].sample(5)
```
