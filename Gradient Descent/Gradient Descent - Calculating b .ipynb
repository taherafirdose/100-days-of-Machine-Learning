```python


import numpy as np
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python

X,y = make_regression(n_samples=10, n_features=1, n_informative=1, n_targets=1,noise=80,random_state=91)
```

```python
plt.scatter(X,y)
```

```python
lr=LinearRegression()
lr.fit(X,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X, lr.predict(X),color='red')
```

```python
# Lets apply Gradient Descent assuming slope is constant m = 79.74
# and let's assume the starting value for intercept b = 0
#y_pred=mx+b
y_pred = ((79.74 * X) + 0)
```

```python
plt.scatter(X,y)
plt.plot(X, lr.predict(X),color='red',label='OLS')
plt.plot(X, y_pred,color='green', label='b is 0')
plt.legend()
plt.show()
```

```python
m = 79.74
b= 0

slope= -2 * np.sum(y - m*X.ravel() - b)

print(slope)
```

```python
#stepsize

learningrate=0.01

stepsize= learningrate* slope
print(stepsize)
```

```python
#calculate the new b

bnew = b - stepsize
print(bnew)
```

```python

y_pred1 = ((79.74 * X) + bnew)
```

```python
plt.scatter(X,y)
plt.plot(X, lr.predict(X),color='green',label='OLS')
plt.plot(X,y_pred1,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()

```

```python
#Iteration 2

m = 79.74
b= -9.345

slope1= -2 * np.sum(y - m*X.ravel() - b)

print(slope1)
```

```python
#stepsize

learningrate=0.01

stepsize1= learningrate* slope1
print(stepsize1)
```

```python
bnew1 = b - stepsize1
print(bnew1)
```

```python

y_pred2 = ((79.74 * X) + bnew1)
```

```python



plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python

b = 0
m = 79.74
lr = 0.01

epochs = 50

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*X.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = m * X + b
  print(b)

  plt.plot(X,y_pred)

plt.scatter(X,y)
```
