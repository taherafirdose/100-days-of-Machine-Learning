What is Dataframe

A dataframe is a two-dimensional labeled data structure that is commonly used for data manipulation and analysis in Python. It is a table-like structure in which the rows represent observations and the columns represent variables or features.

```python
import pandas as pd

#create dataframe using dictionay
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'City': ['New York', 'London', 'Paris']}

df = pd.DataFrame(data)

print(df)
```
