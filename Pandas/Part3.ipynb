```python
#Vectorized String


# Load the Titanic dataset
titanic_df = pd.read_csv('Titanic.csv')
```

```python
titanic_df.head()
```

```python
#Split the name to First name and Last Name

titanic_df['FirstName'] = titanic_df['Name'].str.split(',').str.get(0)

titanic_df.head()
```

```python
titanic_df['LastName']=titanic_df['Name'].str.split('.').str.get(1)
```

```python
titanic_df.head()
```

```python
titanic_df['FullName']=titanic_df['FirstName'].str.cat(titanic_df['LastName'], sep=' ')
titanic_df.head()
```

```python
#Substring
titanic_df['title'] = titanic_df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)

titanic_df.head()
```

```python
# View the unique titles and their counts
print(titanic_df['title'].value_counts())
```

```python
titanic_df.sample(5)
```

```python
# Extracting the ticket number from the Ticket column
titanic_df['Ticket Number'] = titanic_df['Ticket'].str.extract('(\d+)')


titanic_df.sample(5)
```

```python
titanic_df.sample(5)
```

```python
titanic_df['title']=titanic_df['title'].str.replace('Ms','Miss')

print(titanic_df['title'].value_counts())
```

```python
# Filtering for rows where the Name column contains 'Mr.', 'Mrs.', or 'Miss.'
titanic_df[titanic_df['Name'].str.contains('Mr.|Mrs.|Miss.')]['title'].value_counts()


```

```python
# Filter out all the passengers whose name starts with "B" and ends with "e"
titanic_df[titanic_df['Name'].str.startswith('B') & titanic_df['Name'].str.endswith('e', na=False)].head()


```

```python
# Extract the first 3 characters of each name
titanic_df['first_three_chars'] = titanic_df['Name'].str[:3]

titanic_df.sample(5)
```

```python
# Extract the last 5 characters of each name


titanic_df['last_five_chars ']=titanic_df['Name'].str[-5:]
titanic_df.sample(5)
```

```python
# Reverse each name

titanic_df['reversed_names'] = titanic_df['Name'].str[::-1]
titanic_df.sample(5)
```

```python
titanic_df.sample(5)
```

```python
# Convert all names to lowercase
titanic_df['lower']  = titanic_df['Name'].str.lower()
titanic_df.sample(5)
```

```python
titanic_df['upper']  = titanic_df['FirstName'].str.upper()
titanic_df.sample(5)
```

```python
# Capitalize the first letter of each name
titanic_df['capitalize']  = titanic_df['FirstName'].str.capitalize()
titanic_df.sample(5)
```

```python
titanic_df['FullName']=titanic_df['FullName'].str.lower()
```

```python
# Title case each name (capitalize the first letter of each word)
titanic_df['titlecased_names'] = titanic_df['FullName'].str.title()
titanic_df.sample(5)
```
