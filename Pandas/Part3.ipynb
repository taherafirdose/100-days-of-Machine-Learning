```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#load the Titanic dataset:

titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.pivot_table(index='sex', columns=['class'],values='survived')
```

```python
titanic.pivot_table(values=['fare', 'survived'], index='sex', columns='class')
```

```python
titanic.pivot_table(values='fare', index='sex', columns='class', aggfunc=['max', 'min'])
```

```python
titanic.pivot_table(values='survived', index='sex', columns='class', margins=True)
```
