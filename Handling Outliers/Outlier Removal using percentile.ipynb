```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('weight-height.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['Height'])

plt.subplot(2,2,2)
sns.boxplot(df['Height'])

plt.subplot(2,2,3)
sns.distplot(df['Weight'])

plt.subplot(2,2,4)
sns.boxplot(df['Weight'])

plt.show()
```

```python
upper_limit = df['Height'].quantile(0.99)
lower_limit = df['Height'].quantile(0.01)
print('The upper limit', upper_limit)
print('The lower limit', lower_limit)
```

```python
df[(df['Height'] < upper_limit) & (df['Height'] > lower_limit)]
```

```python
df[(df['Height'] > upper_limit)]
```

```python
df[(df['Height'] < lower_limit)]
```

```python
# Capping
cap_df=df.copy()
cap_df['Height'] = np.where(cap_df['Height'] >= upper_limit,
        upper_limit,
        np.where(cap_df['Height'] <= lower_limit,
        lower_limit,
        cap_df['Height']))
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['Height'])

plt.subplot(2,2,2)
sns.boxplot(df['Height'])

plt.subplot(2,2,3)
sns.distplot(cap_df['Height'])

plt.subplot(2,2,4)
sns.boxplot(cap_df['Height'])

plt.show()
```
