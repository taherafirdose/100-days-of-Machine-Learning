```python

print('The min value in placement_exam_marks column', df['placement_exam_marks'].min())
print('The max value in placement_exam_marks column', df['placement_exam_marks'].max())
print('The max value in placement_exam_marks column', df['placement_exam_marks'].mean())
print('The max value in placement_exam_marks column', df['placement_exam_marks'].std())
```

```python
percentile25=df['placement_exam_marks'].quantile(0.25)
print(percentile25)
```

```python
percentile75=df['placement_exam_marks'].quantile(0.75)
print(percentile25)
```

```python
iqr= percentile75 - percentile25
print(iqr)
```

```python
upperlim= percentile25 + 1.5*iqr
lowerlim= percentile25 - 1.5*iqr

print('Upper limit is :', upperlim)
print('Lower limit is :', lowerlim)
```

```python
df[df['placement_exam_marks'] > upperlim]
```

```python
df[df['placement_exam_marks'] < lowerlim]
```

```python
#Trimming

new_df= df[(df['placement_exam_marks'] <= upperlim) & (df['placement_exam_marks'] >= lowerlim)]
new_df
```

```python
# Comparing the plots

plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['placement_exam_marks'])

plt.subplot(2,2,2)
sns.boxplot(df['placement_exam_marks'])

plt.subplot(2,2,3)
sns.distplot(new_df['placement_exam_marks'])

plt.subplot(2,2,4)
sns.boxplot(new_df['placement_exam_marks'])

plt.show()
```

```python
#Capping

cap_df=df.copy()
cap_df['placement_exam_marks']=np.where(df['placement_exam_marks'] > upperlim, upperlim,
                np.where(df['placement_exam_marks'] < lowerlim, lowerlim,
                 df['placement_exam_marks']        
                        ))
```
